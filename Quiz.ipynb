Load smiley

```python
import cv2 as cv
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import numpy as np

path = 'kuis.jpg'

img = cv.imread(path)
```

```python
gray = cv.imread(path, cv.IMREAD_GRAYSCALE)
```

```python
blue = gray.copy()
# blue[:,:,0] = blue[:,:,1] = 0
plt.imshow(blue, cmap='Blues')
```

```python
red = gray.copy()
# red[:,:,1] = red[:,:,2] = 0
plt.imshow(red, cmap='Reds')
```

```python
green = gray.copy()
# green[:,:,0] = green[:,:,2] = 0
plt.imshow(green, cmap='Greens')
```

Ubah Ukuran

```python
resize = img.copy()
resize = cv.resize(resize, (1000, 1000))
plt.imshow(resize)
```

Buat Garis

```python
line = img.copy()

height = line.shape[0]
width = line.shape[1]

cv.line(line, (0, int(height/2)), (width, int(height/2)), (127, 127, 127))
cv.line(line, (int(width/2), 0), (int(width/2), height), (127, 127, 127))

plt.imshow(line)
```

Flip

```python
flip = img.copy()
vertical = cv.flip(flip, -1)
horizontal = cv.flip(flip, 1)

final = cv.hconcat((vertical, horizontal))

plt.imshow(final)
```

Add shapes (that looks like a glass)

```python
glass = img.copy()

cv.circle(glass, center=(85,100), radius=25, color=(255,0,0), thickness=5)
cv.line(glass, (0, 100), (60, 100), (255, 255, 255), thickness=5)

cv.circle(glass, center=(140,100), radius=25, color=(255,127,0), thickness=5)
cv.line(glass, (width, 100), (width-60, 100), (255, 255, 255), thickness=5)

plt.imshow(glass)
```
