**Step 1**
Import libraries

```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

**Step 2 - 4**
Reproduce images, get sizes, add lines horizontally, vertically, and diagonally

```python
urls = [
    "https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
    "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
    "https://placekitten.com/800/571"
]

for url in urls:
  image = io.imread(url)
  image = cv.resize(image, (0,0), fx=.5, fy=.5)

  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
  image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
  image_4 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
  # image_5 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

  height = image_2.shape[0]
  width = image_2.shape[1]

  # horizontal
  for y in range(width):
    image_3[int(height/2), y] = [255, 255, 255]

  # vertical
  for x in range(height):
    image_4[x, int(width/2)] = [255, 255, 255]
  
  # diagonal
  image_5 = cv.line(image, (0,0), (width, height), (255, 255, 255))
  # for x in range(height):
  #   image_5[x, int(x)] = [255, 255, 255]

  final = cv.hconcat((image, image_2, image_3, image_4, image_5))

  cv2_imshow(final)
  print('Image resolution: h x w = ', height, ' x ', width)
  print('\n\n')
```
