```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt 
import numpy as np

path = '/content/drive/MyDrive/Classroom/PCVK/kitten.jpg'

img = cv.imread(path)
plt.imshow(img)
```

```python
img3 = cv.resize(img2, (512, 1024))
plt.imshow(img3)
```

```python
img4 = cv.flip(img2, 0)

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
ax.imshow(img4)
```

```python
black = np.zeros(shape=(512,512,3), dtype=np.int16)
plt.imshow(black)
```

```python
cv.rectangle(black, pt1=(384,0), pt2=(510,150), color=(0,255,0), thickness=10)
plt.imshow(black)
```

```python
cv.rectangle(black, pt1=(200,200), pt2=(300,300), color=(0,0,255), thickness=15)
plt.imshow(black)
```

```python
cv.circle(black, center=(100,100), radius=50, color=(255,0,0), thickness=8)
plt.imshow(black)
```

```python
cv.circle(black, center=(400,400), radius=50, color=(255,255,0), thickness=-1)
plt.imshow(black)
```

```python
cv.line(black, pt1=(0,0), pt2=(512,512), color=(255,255,255), thickness=5)
plt.imshow(black)
```

```python
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(
  black,
  text='Hello',
  org=(10,500),
  fontFace=font,
  fontScale=4,
  color=(255,255,0),
  thickness=2,
  lineType=cv.LINE_AA
)

cv.line(black, pt1=(0, 375), pt2=(512, 375), color=(255,255,255), thickness=2)

plt.imshow(black)
```

```python
black2 = np.zeros(shape=(512, 512, 3), dtype=np.int32)
plt.imshow(black2)
```

```python
vertices = np.array([
    [100,300],
    [200,200],
    [400,300],
    [200,400],
], dtype=np.int32)
vertices
```

```python
pts = vertices.reshape((-1,1,2))
pts
```

```python
cv.polylines(black2, [pts], isClosed=True, color=(255,0,0), thickness=5)
plt.imshow(black2)
```
